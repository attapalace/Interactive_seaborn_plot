```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
import numpy as np
%matplotlib inline

from ipywidgets import interactive,interact
```

```python
df = pd.read_csv('phds_country.csv')
df.head()
```

```python
df.columns
```

```python
ALL = 'All Fields'
phd_columns = ['All Fields','All S&E',
       'Physical & Biological Sciences',
       'Mathematics & Computer Sciences',
    'Agricultural Sciences',
       'Social & Behavioral Sciences',
       'Engineering', 'Non-S&E']
```

```python
dropdown = widgets.Dropdown(
    options=phd_columns,
    value='All Fields',
    description='View Type: ',
    disabled=False,
)

m= widgets.Checkbox(
    value=True,
    description='Male',
    disabled=False,
    indent=False)
f = widgets.Checkbox(
    value=True,
    description='Female',
    disabled=False,
    indent=False)

check_items = [m,f]
items = widgets.GridBox(check_items, layout=widgets.Layout(grid_template_columns="repeat(2, 100px)"))

t = widgets.Label(value="Compare: ")    

first_part = widgets.HBox([dropdown,t,items])

toggle = widgets.ToggleButtons(
    options=['Ascending % of Females', 'Region'],
    description='Order By:',
    disabled=False,
    button_style='',
    value='Ascending % of Females'
)

all_ = widgets.VBox([first_part,toggle])
all_
```

```python
df['total'] = df['All Fields - Female'] + df['All Fields - Male']
```

```python
df_c1 = df.copy()
for col in phd_columns:
    total = df_c1[col + ' - Male'] + df_c1[col + ' - Female']
    df_c1[col + ' - Male'] = df_c1[col + ' - Male'] / total * 100
    df_c1[col + ' - Female'] = df_c1[col + ' - Female'] / total * 100
df_c1.head()
```

```python
def create_plot(phd_type , male, female,order):
    plt.figure(figsize=(14,8))

    if (order == 'Region'):
        df_c1.sort_values('Region',inplace=True)
    else :
        df_c1.sort_values(phd_type + ' - Female',inplace=True)
    
    if male:
        sns.scatterplot(data =df_c1 , x = 'Country', y= phd_type + ' - Male' ,size= df_c1['total']
                            ,sizes=(10,1000),color='orange')
        
    if female:
        sns.scatterplot(data =df_c1 , x = 'Country', y= phd_type + ' - Female' ,size= df_c1['total']
                            ,sizes=(10,1000),color='c')
        
    plt.xticks(rotation=90)
    plt.ylabel('Percent Of Total')
    plt.legend('')
    plt.ylim(0,100)
    plt.show()
    ;

interactive_plot = interactive(create_plot, phd_type=dropdown,male = m, female=f ,order = toggle)
output = interactive_plot.children[-1]
interactive_plot
```
